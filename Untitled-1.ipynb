```python
x_train,x_test,y_train,y_test=train_test_split(x,y,shuffle=True,test_size=0.15,random_state=42)
```

```python
#select model
lei_reg=LinearRegression()
lei_reg.fit(x_train,y_train)
lei_reg.predict(x_test)
```
